```python
df1.head()
```

```python
df1.BULAN = df2.BULAN.str.replace("-", " ")
```

```python
df1.BULAN = pd.to_datetime(df1.BULAN, format="%b %y")
```

```python
month_vals = []
for i in range(df1.shape[0]):
    month_vals.append(i)

df1['URUTAN'] = month_vals
```
